```python
import numpy as np
import matplotlib.pyplot as plt
#Define la funcion para plotear
def plot_function(f, x_range, step=0.01):
    x = np.arange(x_range[0], x_range[1], step)
    y = f(x)

#funciones matplotlib
    plt.plot(x, y, label="f(x)")
    plt.axhline(0, color='k', linewidth=0.8)  # x-axis
    plt.grid(True)
    plt.xlabel("β")
    plt.ylabel("f(β)")
    plt.legend()
    plt.show()

# seccion principal
if __name__ == "__main__":
    # funcion de ejemplo (realmente es la ffinal
    def example_function(x):
        return np.cosh(x)*np.cos(x)-1


    # Define rango para x
    x_range = (4, 8)

    # Plotea
    plot_function(example_function, x_range)
```

```python
def newton_raphson(f, f_prime, x_0, e_s=1e-8, n_max=1000):
    e_a = 1000
    n = 0
    e_a_list = []  # Lista para almacenar el error en cada iteración
    n_list = []  # Lista para almacenar el número de iteraciones

    while e_a > e_s and n < n_max:
        delta_x = 0.001 * x_0
        x = x_0 - f(x_0) / f_prime(x_0, delta_x)
        e_a = abs((x - x_0) / x)
        x_0 = x
        n += 1
        e_a_list.append(e_a)  # Guardar el error de cada iteración
        n_list.append(n)  # Guardar el número de iteración

    if n == n_max:
        return None, e_a_list, n_list  # Si no se encuentra la raíz por el límite de iteraciones
    else:
        return x_0, e_a_list, n_list  # Si se encuentra la raíz, la lista de errores y la lista de iteraciones

# Main script que corre el código
if __name__ == "__main__":
    # Función f(x)
    def f(x):
        return np.cosh(x) * np.cos(x) - 1


    # Derivada aproximada de f(x)
    def f_prime(x, delta_x):
        return (f(x + delta_x) - f(x - delta_x)) / (2 * delta_x)

    x_0 = 4.7  # Valor inicial

    raiz, errores, iteraciones = newton_raphson(f, f_prime, x_0)

    if raiz is None:
        print("Terminado por límite de iteraciones")
    else:
        print('Raíz es:', raiz)
        print('Con un error de aproximación de:', errores[-1])
        print('Y tomó', iteraciones[-1], 'iteraciones')
```
